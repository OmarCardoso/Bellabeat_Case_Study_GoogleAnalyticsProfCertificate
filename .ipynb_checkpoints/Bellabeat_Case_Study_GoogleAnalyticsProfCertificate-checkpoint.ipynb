## Seguem as transformações de dados necessárias
1. Converter ActivityDate para tipo datatime64<br>
2. Converter formato Activitydate para dd-mm-YYYY<br>
3. Criar nova coluna "DiaSemana" separando o dia da semana para análises futuras<br>
4. Criar nova coluna "TotalMinutos" que será a soma de VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes and SedentaryMinutes<br>
5. Criar nova coluna "TotalHoras" que será a conversão da nova coluna 4 para o número de horas<br>
6. Renomear as colunas para um melhor entendimento

```python
# 1 - Converter ActivityDate para tipo datatime64 
atividade_diaria["ActivityDate"] = pd.to_datetime(atividade_diaria["ActivityDate"])
# format="%d/%m/%Y")                                            
# verificar o erro ao colocar format -> ValueError: time data '4/13/2016' does not match format '%d/%m/%Y' (match)   
                                                  
# 2 - df["Data Pagamento"] = pd.to_datetime(df["Data Pagamento"]).dt.strftime('%d-%m-%Y') 
# a função dt.strftime embora ajuste o formato desejado transforma o type para objetct

```

```python
# exibir para checar a conversão
atividade_diaria.info()
# exibir as 5 primeiras linhas para confirmar
atividade_diaria['ActivityDate'].head()
```

Criar nova lista de colunas com novos nomes e renomear o data frame para df_atividades

```python
# 3,4 e 5 - Criar nova lista de colunas incluindo as colunas "DayOfWeek", "TotalExerciseMinutes", "TotalExerciseHours" 
novas_col = ['Id', 'ActivityDate', 'DayOfWeek', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalExerciseMinutes', 'TotalExerciseHours', 'Calories']

# função reindex para rearranjar as colunas baseado em "novas_col"
# foram repetidos os nomes das colunas e incluidas 3 novas colunas
df_atividade = atividade_diaria.reindex(columns=novas_col)

# imprimir as 5 primeiras para verificar
df_atividade.head()
```

```python
# 3 - Preencher a coluna dia da semana para análises futuras
df_atividade['DayOfWeek'] = df_atividade['ActivityDate'].dt.day_name()

# exibe as 5 primeiras linhas para checar
df_atividade['DayOfWeek'].head()
```

```python
6 -Renomear as colunas para um melhor entendimento
estamos antecipando este passo antes de preecher as novas colunas totalizadoras (itens 4 e 5) 
pois, facilitar o manuseio do dataset
Alterando o formato XxxYyy para xxx_yyy e colocando em portugues
```

```python
# renomeando as colunas
df_atividade.rename(columns = {'Id' : 'Id', 'ActivityDate' : 'data_atividade', 'DayOfWeek' : 'dia_semana', 'TotalSteps' : 'total_passos', 'TotalDistance' : 'total_distancia', 'TrackerDistance' : 'rastreador_distancia', 'LoggedActivitiesDistance' : 'apur_dist_atividade', 'VeryActiveDistance' : 'alta_dist_atividade', 'ModeratelyActiveDistance' : 'moderada_dist_atividade', 'LightActiveDistance' : 'leve_dist_atividade', 'SedentaryActiveDistance' : 'sedentaria_dist_atividade', 'VeryActiveMinutes' : 'alta_min_atividade', 'FairlyActiveMinutes' : 'medio_min_atividade', 'LightlyActiveMinutes' : 'leve_min_atividade', 'SedentaryMinutes' : 'sedentario_min_atividade', 'TotalExerciseMinutes' : 'total_exercicio_min', 'TotalExerciseHours' : 'total_exercicio_horas', 'Calories' : 'calorias'}, inplace = True)

# Exibe colunas para checar
print(df_atividade.columns.values)
df_atividade.head(5)
```

```python
# 4 - Preencher a nova coluna total_exercicio_min que será a soma do total de minutos
df_atividade['total_exercicio_min'] = df_atividade['alta_min_atividade'] + df_atividade['medio_min_atividade'] + df_atividade['leve_min_atividade'] + df_atividade['sedentario_min_atividade']

# exibe para checar
df_atividade['total_exercicio_min'].head()
```

```python
# 5 - Preencher a nova coluna total_exercicio_horas que será a conversao do total de minutos para horas
df_atividade['total_exercicio_horas'] = round(df_atividade['total_exercicio_min'] / 60)

# exibir para checar
df_atividade['total_exercicio_horas'].head()
```

## 5.2 Análise Bivariada

```python
# plotando scatter plot (1)
plt.style.use('default')
plt.figure(figsize=(8,6))
plt.scatter(df_analise.total_passos, df_analise.calorias, alpha = 0.8, c = df_analise.calorias, cmap = "Spectral")

# adiciona anoataçãoes e visual
media_calorias = 2303
media_passos = 7637

plt.colorbar (orientation = "vertical")
plt.axvline(media_passos, color = 'Blue', label = "Media Passos")
plt.axhline(media_calorias, color = 'Red', label = 'Media de calorias queimadas')
plt.xlabel("Passos percorridos")
plt.ylabel("Calorias queimadas para cada passo")
plt.grid(True)
plt.legend()
plt.show
```

Com o Scatter plot ficou identificado:<br>
    1. Existe uma correlação positiva<br>
    2. Observamos que a "intensidade" de calorias queimadas aumenta quando as usuárias estão entre 0 e 15.000 passos.<br>
    A partir de 15.000 passos, a taxa de calorias queimadas diminui drasticamente.<br>
    3. Identificamos alguns outliers<br>
    - Zero passos com zero para as mínimas calorias queimadas<br>
    - Estes outliers podem ser devido a variação natural dos dados, mudança no uso das usuárias ou erro no dataset<br>
      (erro de cálculo, contaminação de dados ou erro humano)

```python
# plotando scatter plot (2)
plt.style.use('default')
plt.figure(figsize = (8,6))
plt.scatter(df_analise.total_exercicio_horas, df_analise.calorias, alpha = 0.8, c = df_analise.calorias, cmap = 'Spectral')

# adicionando anotações e visual
media_calorias = 2303
media_horas = 20
media_sedentaria = 991 / 60

plt.colorbar(orientation = 'vertical')
plt.axvline(media_horas, color = 'Blue', label = 'Média de horas registradas')
plt.axvline(media_sedentaria, color = 'Purple', label = 'Media sedentária')
plt.axhline(media_calorias, color = 'Red', label = 'Media calorias')
plt.xlabel('Horas registradas')
plt.ylabel('Calorias queimadas')
plt.title('Calorias queimadas para cada hora registrada')
plt.legend()
plt.grid(True)
plt.show
```

Com este scatter plot identificamos:
    1 - A fraca correlação em que o aumento de horas registradas não traduz em mais calorias queimadas se deve em grande parte 
à média de horas sedentárias (linha roxa) plotada no intervalo de 16 e 17 horas.<br>
    2 - Novamente, podemos ver alguns outliers:<br>
        - O mesmo valor zero outliers<br>
        - Um ponto zero em 24 horas que pode ser devido as mesmas razões da plotagem acima.
        
        

```python
# Exibindo gráfico de pizza
# Calculo total de minutos individuais
alta_atividade_min = df_atividade['alta_min_atividade'].sum()
media_atividade_min = df_atividade['medio_min_atividade'].sum()
leve_atividade_min = df_atividade['leve_min_atividade'].sum()
sedentaria_atividade_min = df_atividade['sedentario_min_atividade'].sum()

# plotando gráfica de pizza
slices = [alta_atividade_min, media_atividade_min, leve_atividade_min, sedentaria_atividade_min]
labels = ['alta_atividade_min', 'meida_atividade_min', 'leve_atividade_min', 'sedentaria_atividade_min']
colours = ['lightcoral', 'yellowgreen', 'lightskyblue', 'darkorange']
explode = [0, 0, 0, 0.1]
plt.pie(slices, labels = labels, 
        colors = colours, wedgeprops = {"edgecolor": "black"}, 
        explode = explode, autopct = "%1.1f%%")
plt.title('Percentual de atividades em minutos')
plt.tight_layout()
plt.show()
```

Como vimos no gráfico de pizza

1. Minutos sedentários representam a maior fatia, 81,3%
2. Isto indica que as usuárias do FitBit App estão usando o app para registrar atividades diárias como deslocamento,
movimentos inativos, (movendo de um local para outro).
3. Pelo baixo percentual de atividade física média e alta, o FitBit App esta sendo pouco usado para registrar
este tipo de atividade (ex corrida). Isso é altamente desencorajador, pois o aplicativo FitBit foi desenvolvido
para incentivar o condicionamento físico.
